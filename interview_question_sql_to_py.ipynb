```python
df['rn']=df.sort_values(['timedate','Name'],ascending=[False,True]).groupby(['Name']).cumcount()+1
## groupby.cumcount() is used to add a sequential counter column on groups and +1 is used if we want orderings starting at 1.


dfN=df[df.rn==1]
dfN=dfN.drop(['rn'],axis=1)
# Make Own Index as index
# In this case default index still exist 
#dfN.reset_index(inplace = True)
#dfN

# replacing own index with default index
dfN.reset_index(drop=True,inplace=True)
dfN
```
